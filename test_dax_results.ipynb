```python
import pandas as pd
import numpy as np
```

```python
match_home = pd.read_csv('Data/match_home_team.csv')
match_home
```

```python
match_away = pd.read_csv('Data/match_away_team.csv')
match_away
```

```python
match = pd.merge(match_home, match_away[['match_id', 'away_team_id', 'away_team_goal']], on='match_id', how='left')
match
```

```python
team = pd.read_csv('Data/team.csv')
```

```python
team[team['team_long_name'].str.contains('Brugge')]
```

```python
brugge = match[((match['home_team_id']==8342) | (match['away_team_id']==8342)) & (match['season_id']==7)]
brugge
```

```python
goals_for = brugge[brugge['home_team_id']==8342]['home_team_goal'].sum() + brugge[brugge['away_team_id']==8342]['away_team_goal'].sum()
goals_for
```

```python
goals_against = brugge[brugge['home_team_id']==8342]['away_team_goal'].sum() + brugge[brugge['away_team_id']==8342]['home_team_goal'].sum()
goals_against
```

```python
win = (
       len(brugge[(brugge['home_team_id']==8342) & (brugge['home_team_goal'] > brugge['away_team_goal'])]) 
       + len(brugge[(brugge['away_team_id']==8342) & (brugge['away_team_goal'] > brugge['home_team_goal'])]) 
      )
win
```

```python
draw = (
        len(brugge[(brugge['home_team_id']==8342) & (brugge['home_team_goal'] == brugge['away_team_goal'])]) 
        + len(brugge[(brugge['away_team_id']==8342) & (brugge['away_team_goal'] == brugge['home_team_goal'])]) 
       )
draw
```

```python
loss = (
        len(brugge[(brugge['home_team_id']==8342) & (brugge['home_team_goal'] < brugge['away_team_goal'])]) 
        + len(brugge[(brugge['away_team_id']==8342) & (brugge['away_team_goal'] < brugge['home_team_goal'])]) 
       )
loss
```

```python
points = (win * 3) + draw
points
```
